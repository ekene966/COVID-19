##Midterm Exam: 
Design a python program to draw a wordcloud on this notebook. (1) The wordcloud must show only the titles of movies in Netflix. For input data, use netflix_titles.csv (80 points). (2) The size of the wordcloud is 20x10 (10 points). (3) The title of the wordcloud is "Titles of Movies in Netflex" (10 points).

Note: netflix_titles.csv is in http://csit.udc.edu/~jkim/data/

```
netflix = pd.read_csv('netflix_titles.csv')
netflix.head()
```

```
netflix.info()
```

```
movies = netflix[netflix.type == 'Movie']
shows = netflix[netflix.type == 'TV Show']
```

```

import matplotlib.pyplot as plt
def plot_cloud(wordcloud):
  plt.figure(figsize=(20,10))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.title("Titles of Movies in Netflix", size='20')

text =' '.join(movies['title'])

wordcloud = WordCloud(width = 500, height = 500, background_color='#40E0D0', colormap="ocean",  random_state=10).generate(text)
plot_cloud(wordcloud)
```
